````python
#SECTION - 1 : Dogs breed modules 

# This module contains the DogBreed class which is used to create a dog breed object


# Imports
import typing
temperamentLevels = typing.Literal["Friendly", "Neutral", "Agressive"]


class DogBreed:
    """
    This class is used to create a dog breed object
    ```python
    dog =  new DogBreed("Labrador", 24, 30, 12, "Friendly", False, "Canada", "Sporting")
    ```
    
    @Attributes:
    ----------
    breed : str
        Breed name of the dog
    height : float
        Height of the dog
    weight : float
        Weight of the dog
    life_span : int
        Life span of the dog
    temperament : temperamentLevels
        Temperament of the dog (Friendly, Neutral, Aggressive)
    IsWorkDog : bool
        If the dog is a working dog or not (default is False)
    origin : str
        Origin of the dog
    group : str
        Group of the dog
    
    @Methods:
    -------
    get_breed():
        Returns the breed of the dog
    get_height():
        Returns the height of the dog
    get_weight():
        Returns the weight of the dog
    get_life_span():
        Returns the life span of the dog
    get_temperament():
        Returns the temperament of the dog
    get_origin():
        Returns the origin of the dog
    get_group():
        Returns the group of the dog
    __str__():
        Returns the string representation of the dog breed object
        ```python
        dog = DogBreed("Labrador", 24, 30, 12, "Friendly", False, "Canada", "Sporting")
        print(dog)
        ```
    """
    
    
    def __init__(self, breed, height, weight, life_span, temperament:temperamentLevels,IsWorkDog:False, origin, group):
        self.breed = breed
        self.height = height
        self.weight = weight
        
        if life_span <= 0: #info grabbed from source:https://www.petmd.com/dog/care/longest-living-dog-breeds
            raise ValueError("Life span should be a positive number")
        elif life_span >30:
            raise ValueError("Life span should be less than 30 (Dogs don\'t live that long)")
        else:
            self.life_span = life_span
        
        if temperament not in typing.get_args(temperamentLevels):
            raise ValueError("Temperament should be either Friendly, Neutral or Agressive")
        else: 
            self.temperament = temperament
            
        self.IsWorkDog = IsWorkDog
        self.origin = origin
        self.group = group
        
        
    def get_breed(self):
        """
        This method returns the breed of the dog
        """
        return self.breed
    
    def get_height(self):
        """
        This method returns the height of the dog
        """
        return self.height
    
    def get_weight(self):
        """
        This method returns the weight of the dog
        """
        return self.weight
    
    def get_life_span(self):
        """
        This method returns the life span of the dog
        """
        return self.life_span
    
    def get_temperament(self):
        """
        This method returns the temperament of the dog
        """
        return self.temperament
    
    def get_origin(self):
        return self.origin
    
    def get_group(self):
        return self.group
        
    def __str__(self):
        return f"Breed:{self.breed}\nHeight:{self.height}\nWeight:{self.weight}\nLife Span:{self.life_span}\nTemperament:{self.temperament}\nIs a workign dog:{self.IsWorkDog}\nOrigin:{self.origin}\nGroup:{self.group}"
    # End of DogBreed class


        
                

````

````python
#SECTION - 2 : dog database program

#imports
import time #importing the time module for the sleep function
from __main__ import DogBreed #importing the DogBreed class from the DogBreed module

dogs = []
def startup():
    """
    This function is used to start the program
    """
    print("Welcome to the Dog Database Program")
    print("This program will help you to store and retrieve information about dog breeds")
    print("Let's get started")
    print("\n\n")
    mainMenu()
    
#addDogBreed function is used to add a dog breed to the database
def addDogBreed():
    """
    This function is used to add a dog breed to the database
    """
    breed = input("Enter the breed of the dog: ")
    height = float(input("Enter the height of the dog: "))
    weight = float(input("Enter the weight of the dog: "))
    
    life_span = int(input("Enter the life span of the dog: "))
    if life_span <= 0 or life_span > 30:
        print("Life span should be a positive number")
        life_span = int(input("Enter the life span of the dog: "))
    
    temperament = input("Enter the temperament of the dog (Friendly, Neutral, Aggressive): ")
    if temperament not in ["Friendly", "Neutral", "Aggressive"]:
        print("Temperament should be either Friendly, Neutral or Aggressive")
        temperament = input("Enter the temperament of the dog (Friendly, Neutral, Aggressive): ")
    
    IsWorkDog = input("Is the dog a working dog (True/False): ")
    if IsWorkDog not in ["True", "False"]:
        print("IsWorkDog should be either True or False")
        IsWorkDog = input("Is the dog a working dog (True/False): ")
    IsWorkDog = IsWorkDog == "True"
    origin = input("Enter the origin of the dog (e.g: Manchester New Hampshire): ")
    
    group = input("Enter the group of the dog (e.g: Sporting, Hound, Working, Terrier, Toy, Non-Sporting, Herding or Miscellaneous ): ")
    if group not in ["Sporting", "Hound", "Working", "Terrier", "Toy", "Non-Sporting", "Herding", "Miscellaneous"]: #info grabbed from source:https://www.akc.org/dog-breeds/groups/
        print("Group should be either Sporting, Hound, Working, Terrier, Toy, Non-Sporting, Herding or Miscellaneous")
        group = input("Enter the group of the dog (e.g: ): ")
    try:
        dogs.append(DogBreed(breed, height, weight, life_span, temperament, IsWorkDog, origin, group))
        print("Dog breed added successfully")
    except ValueError as e:
        print(e)
    mainMenu() 
    
#displayAllDogBreeds function is used to display all the dog breeds in the database
def displayAllDogBreeds():
    """
    This function is used to display all the dog breeds in the database
    """
    if len(dogs) == 0:
        print("No dog breeds found")
    else:
        for dog in dogs:
            print(dog)
            print("\n")
            
    input("Press enter to return to the main menu")
    
    mainMenu()

#searchDogBreed function is used to search for a dog breed in the database
def searchDogBreed():
    """
    This function is used to search for a dog breed in the database
    """
    breed = input("Enter the breed of the dog you want to search for: ")
    found = False
    for dog in dogs:
        if breed.lower() in dog.get_breed().lower():
            print(f"{dog}\n")
            found = True
    if not found:
        print("Dog breed not found")
    mainMenu()
   
#mainMenu function is used to display the main menu of the program
def mainMenu():
    """
    This function is used to display the main menu of the program
    """
    print("Loading Main Menu...")
    time.sleep(3)
    print("\n\n")
    print("Main Menu")
    print("1. Add a dog breed")
    print("2. Display all dog breeds")
    print("3. Search for a dog breed")
    print("4. Load Sample Data")
    print("5. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        addDogBreed()
    elif choice == 2:
        displayAllDogBreeds()
    elif choice == 3:
        searchDogBreed()
    elif choice == 4:
        print("Loading Sample Data...")
        dogs.append(DogBreed("Labrador", 24, 30, 12, "Friendly", False, "Canada", "Sporting"))
        print("Labrador breed added successfully")
        dogs.append(DogBreed("Golden Retriever", 24, 35, 12, "Friendly", False, "United Kingdom", "Sporting"))
        print("Golden Retriever breed added successfully")
        
        #add two german shepherds with different temperaments to show the search function works
        dogs.append(DogBreed("German Shepherd", 24, 40, 13, "Friendly", True, "Germany", "Herding"))
        dogs.append(DogBreed("German Shepherd", 29, 40, 13, "Agressive", True, "Germany", "Herding"))
        print("German Shepherd breed added successfully")
        dogs.append(DogBreed("Bulldog", 14, 50, 8, "Friendly", False, "United Kingdom", "Non-Sporting"))
        print("Bulldog breed added successfully")
        dogs.append(DogBreed("Poodle", 15, 70, 15, "Friendly", False, "Germany", "Non-Sporting"))
        print("Poodle breed added successfully")
        print("Sample data loaded successfully")
        mainMenu()
        
    elif choice == 5:
        print("Thank you for using the program")
        print("Goodbye")
    else:
        print("Invalid choice. Please try again")
        mainMenu()
        
startup() #calling the startup function to start the program
#End of the program
````

### Question 1: What is the significance of indentation in Python?

Indentation in Python is crucial because it defines the structure and scope of code blocks. Unlike many other programming languages that use braces or keywords to delimit blocks, Python uses indentation. ### Question 2: How does Python handle comments in code?

Python handles comments using the `#` symbol for single-line comments. Any text following `#` on a line is ignored by the interpreter. Multi-line comments can be created using triple quotes (`'''` or `"""`) to enclose the comment block.

### Question 8: What is the difference between mutable and immutable data types?

Mutable data types can be changed after they are created (e.g., lists, dictionaries, sets). Immutable data types cannot be changed after they are created (e.g., integers, floats, strings, tuples). If you try to modify an immutable object, a new object is created.

### Question 9: How does Python handle dynamic typing?

Python uses dynamic typing, which means that the type of a variable is determined at runtime. You don't need to explicitly declare the type of a variable; Python infers it based on the assigned value.

### Question 11: What are the different types of operators available in Python?

### Question 12: How does Python handle division differently when using `/` vs `//`?

*   `/` performs floating-point division, which always returns a float, even if the operands are integers.
*   `//` performs floor division, which returns the largest whole number that is less than or equal to the result.

```python
print(5 / 2)   # Output: 2.5
print(5 // 2)  # Output: 2
```

### Question 13: What is the purpose of the `in` operator in Python?

The `in` operator is used to check if a value exists within a sequence (e.g., list, tuple, string). It returns `True` if the value is found, and `False` otherwise.

```python
my_list = [1, 2, 3]
print(2 in my_list)  # Output: True
print(4 in my_list)  # Output: False
```

### Question 14: What is the difference between `is` and `==` in Python?

*   `==` checks if the values of two objects are equal.
*   `is` checks if two variables refer to the same object in memory.

```python
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)  # Output: True (values are equal)
print(a is b)  # Output: False (different objects in memory)

c = a
print(a is c)  # Output: True (same object in memory)
```

### Question 15: How does Python handle operator precedence?

2.  Exponentiation
3.  Multiplication and Division
4.  Addition and Subtraction

### Question 17: How do `if`, `elif`, and `else` statements work in Python?

*   `if`: Executes a block of code if a condition is true.
*   `elif`: (else if) Checks an additional condition if the previous `if` or `elif` condition is false.
*   `else`: Executes a block of code if all preceding `if` and `elif` conditions are false.

```python

if x > 0:
    print("Positive")
elif x < 0:
    print("Negative")
else:
    print("Zero")
```

### Question 18: What is the difference between `while` and `for` loops in Python?

*   `for` loop: Iterates over a sequence (e.g., list, tuple, string) or other iterable object, executing a block of code for each item in the sequence.

### Question 20: What does the `continue` statement do in a loop?

### Question 22: What is the purpose of the `return` statement in a function?

The `return` statement is used to exit a function and return a value to the caller. If a function doesn't have a `return` statement, it returns `None` by default.

### Question 23: How can you import a module in Python?

You can import a module in Python using the `import` statement, followed by the module name.

```python

```

### Question 24: What is the difference between importing a module with `import` vs `from module import`?

*   `import module`: Imports the entire module. You need to use the module name to access its contents (e.g., `module.function()`).
*   `from module import`: Imports specific attributes or functions from the module directly into the current namespace. You can use them without the module name (e.g., `function()`).

```python
import math
print(math.sqrt(25))

from math import sqrt
print(sqrt(25))
```
